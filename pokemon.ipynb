```python
import pandas as pd 

poke = pd.read_csv("pokemon.csv") #

poke.head(10)
```

```python
poke[["Name","Speed"]]
```

```python
poke[["Name", "Sp. Atk", "Sp. Def"]].sort_values(["Sp. Atk"], ascending=0)
```

```python
poke["Power"]= poke["Attack"]+poke["Defense"]+poke["HP"]+poke["Sp. Atk"]+poke["Sp. Def"]+poke["Speed"]
poke[["Name", "Generation","Legendary","Power"]].sort_values(["Power"],ascending=0) 
```

```python
poke.loc[poke["Power"]>=500,'Powerful']="True"
poke.loc[poke["Power"]<500,"Powerful"]="False"
```
